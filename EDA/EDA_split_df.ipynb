```python
import pandas as pd
```

```python
df = pd.read_csv('./mrc-level2-nlp-09/test.csv',error_bad_lines=False)
```

```python
df
```

```python
df['context'][0]
```

```python
div_context_df = pd.DataFrame([],columns=["question","id","context_id","context","scores","question_num"])
```

```python
div_context_df
```

```python
from tqdm import tqdm

topk=10
for i in tqdm(range(len(df))):
    id_list = list(map(int,df['context_id'][i][2:-2].split('\', \'')))
    context_list = df.loc[i]['context'].split('[DIV]')
    score_list = list(map(float,df['scores'][i][1:-1].split()))
    
    for j in range(topk):
        div_context_df.loc[len(div_context_df)] = [
            df.loc[i]["question"],
            df.loc[i]["id"],
            id_list[j],
            context_list[j],
            score_list[j],
            i
        ]
```
